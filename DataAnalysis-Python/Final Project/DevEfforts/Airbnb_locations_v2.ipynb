```python
location = a[['host_location','smart_location','host_neighbourhood','street',\
             'neighbourhood_cleansed','city','state','zipcode']].copy()
```

```python
location.head()
```

```python
location.state.loc[location.state=='California'] = 'CA'
location.state.unique()
location.state = location['state'].str.upper()
```

```python
location.state.value_counts()
```

```python
location.drop(location[~(location.state.values=='CA')].index,axis=0,inplace=True)
```

```python
location.city = location.city.str.lower()
location.city.unique()
location.city.dropna(inplace=True)
```

```python
location['city'] = location['city'].str.replace(r'[^\x00-\x7F]+', '')
location.city.dropna(inplace=True)
location.city.value_counts()
```
